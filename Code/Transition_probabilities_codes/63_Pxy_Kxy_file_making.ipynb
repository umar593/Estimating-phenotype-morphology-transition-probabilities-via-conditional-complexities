```python
import numpy as np
from collections import OrderedDict
import random as rm
import time

file_inp1 = open('data/s_28_list.txt',"r")
file_out1 = open('data/binary_rep_init.txt',"w")
file_out2 = open('data/general_info_computations_1.txt',"w")
file_out3 = open('data/binary_rep_1.txt',"w")
file_out4 = open('data/sample_binary_rep_1.txt',"w")
file_out5 = open('data/reverse_indices_1.txt',"w")
file_out6 = open('data/reverse_output_1.txt',"w")

sample_size = 10000
arr1 = file_inp1.read()
arr_t1 = arr1.split("\n")
arr_t1.pop()


#print(arr_t1)
print(len(arr_t1))
print("File Reading Done.")

binary_repr_gen = []    
for i in range(0,len(arr_t1)):
    temp = np.binary_repr(i,24)
    binary_repr_gen.append(temp)
    file_out1.write(temp)
    file_out1.write("\n")

#has_duplicates = len(binary_repr_gen) != len(set(binary_repr_gen))

#if has_duplicates:
#    print("The list has duplicate entries.")
#else:
#    print("The list has no duplicate entries.")    
    
#print(binary_repr_gen)
#print(len(binary_repr_gen))
print("Done with general representation and file writing.")

print("time calculation started.")
start_time = time.time()
target_num = '1'

print("The targeted number is",target_num)
file_out2.write("General info \n")
file_out2.write("The number selected is \n")
file_out2.write(target_num)
file_out2.write("\n")

indices = []
for index, value in enumerate(arr_t1):
    if value == target_num:
        indices.append(index)

file_out2.write("The length of indices for target number is \n")
file_out2.write(str(len(indices)))
file_out2.write("\n")        
print("The length of indices for targeted number is ",len(indices))
file_out2.write("The indices for target number is \n")
file_out2.write(str(indices))
file_out2.write("\n")

#if len(indices) > 0:
#    print("Number:", target_num)
#    print("Indices:", indices)
#else:
#    print("Number not found in the list.")  
print("Indices found for targeted number")

binary_repr = []    
for i in range(0,len(indices)):
    num_temp = indices[i]
    temp = np.binary_repr(num_temp,24)
    binary_repr.append(temp)
    file_out3.write(str(temp))
    file_out3.write("\n")

#print(binary_repr)
#print(len(binary_repr))
print("Binary representations done for the indices")

new_sample = rm.sample(binary_repr, sample_size)
print(new_sample)
for i in range(0,len(new_sample)):
    file_out4.write(str(new_sample[i]))
    file_out4.write("\n")
print("New Random Sampling done")
new_col_1 = []

for i in range(0,len(new_sample)):
    x = new_sample[i]
    index = 0
    for j in range(0,len(x)):
        temp = list(x)
        if x[index] == "0":
            temp[index] = "1"
            x1 = "".join(temp)
            new_col_1.append(x1)
        else:
            temp[index] = "0"
            x1 = "".join(temp)
            new_col_1.append(x1)
        index = index + 1


#print(new_col_1)
#print(len(new_col_1))

print("new list generated from saples by flipping.")

reverse_indices = []
for i in range(0,len(new_col_1)):
    try:
        index = binary_repr_gen.index(new_col_1[i])
        reverse_indices.append(index)
        file_out5.write(str(index))
        file_out5.write("\n")
    except ValueError:
        reverse_indices.append("000")
        file_out5.write("000")
        file_out5.write("\n")
    print("Done for string ",i,"out of ",len(new_col_1))
print("Reverse indices are:", reverse_indices)
print("Reverse indices found")

reverse_output = []    
for i in range(0,len(reverse_indices)):
    num_temp = reverse_indices[i]
    temp = arr_t1[num_temp]
    reverse_output.append(temp)
    file_out6.write(str(temp))
    file_out6.write("\n")

print("Reverse output are:", reverse_output)
print("Reverse output found")


end_time = time.time()

print("Total time in seconds is",end_time-start_time)

file_out1.close()    
file_out2.close() 
file_out3.close()
file_out4.close()
file_out5.close()
file_out6.close()
print("Bing0o0oo0o0o0o0o0o0o0")
```
