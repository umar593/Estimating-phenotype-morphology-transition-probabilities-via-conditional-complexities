```python
import numpy as np
from collections import OrderedDict
import random as rm
#from ttictoc import tic,toc
#t1 = tic()

file = open("comstr4.txt", "r")
file1 = open("unique_str_count_prob.txt","w")
file2 = open("unique_str.txt","w")
file3 = open("unique_str_prob.txt","w")

arr = file.read()
arr_t = arr.split("\n")
arr_t.pop()

#print(arr_t)

str_size = 4  # size of the string
col_1 = []
col_2 = []
col_3 = []

#temp = arr_t[0]
#print(len(temp))
#print(arr_t[0])
#print("Printed")

for i in range (0,len(arr_t)):
    temp= arr_t[i]
    temp1 = temp[0:str_size]
    col_1.append(temp1)
    temp2 = temp[str_size+1:2*str_size+1]
    col_2.append(temp2)
    temp3 = temp[2*str_size+2:]
    col_3.append(temp3)

len_in_out = len(col_1) 
#print("Reading and splitting done")
#print("sample ", col_1[0], col_2[0], col_3[0])
#print("Inputs")
#print("---------------------------------")
#print(col_1)
#print("---------------------------------")
#print("Outputs")
#print("---------------------------------")
#print(col_2)
#print("---------------------------------")
#print("frequency")
#print("---------------------------------")
#print(col_3)
#print("---------------------------------")

arrlist = np.array(col_2)
unique, counts = np.unique(arrlist, return_counts=True)
w = dict(zip(unique, counts))
#print(w)

total_outputs = 0
for kv in w.items():
    total_outputs = total_outputs+kv[1]
#print("Total sum is ",total_outputs)
    
prob_set = []
for kv in w.items():
    prob_ind = kv[1]/total_outputs
    prob_set.append(prob_ind)
#print(prob_set)
    
for kv in w.items():
    file1.write(str(kv[0]))
    file2.write(str(kv[0]))
    file1.write('  ')
    file1.write(str(kv[1]))
    file1.write('  ')
    file1.write(str(kv[1]/total_outputs))
    file3.write(str(kv[1]/total_outputs))
    file1.write("\n")
    file2.write("\n")
    file3.write("\n")
print("Bingo")


file.close()
file1.close()
file2.close()
file3.close()

#t2 = toc()

#print(t2)
```
