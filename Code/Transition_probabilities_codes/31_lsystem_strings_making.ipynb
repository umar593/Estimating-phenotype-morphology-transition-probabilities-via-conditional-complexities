```python
import pandas as pd
import numpy as np
import random as rm

file1 = open("In_Out_New_string.txt","w")
file2 = open("Mapped_Output_string.txt","w")
file3 = open("Mapped_Output_prob_string.txt","w")
file4 = open("Mapped_Input_string.txt","w")

data = pd.read_csv('lsystems.csv')
#print(data)

logFreq = data['logFreq'].tolist()
KC = data['KC'].tolist()
geno = data['geno'].tolist()
phenoBinary = data['phenoBinary'].tolist()
phenoLength = data['phenoLength'].tolist()

#print('log freq:',logFreq)
#print(KC[4])
#print(geno[100])
#print(phenoBinary[4])
#print(phenoLength[4])

ind = rm.randint(0,len(geno))
print("Randomly selected string index is", ind)
print("Randomly selected string is ", geno[ind], phenoBinary[ind], phenoLength[ind])
file4.write(str(phenoBinary[ind]))

possible_strs = ['[',']','+','-','F']
#print(possible_strs)
sample_str = geno[ind]
print("The sample string is: ")
print(sample_str)
print("----------------------------------")
new_input_str = []

new_replace_str = []
new_delete_str = []
new_add_str = []

def str_replacement(new_str, possible_strs):
    new_str_l = list(new_str)
    for l in range(len(new_str_l)):
        new_str_l = list(new_str)
        str_ele = new_str_l[l]
        for m in range(0,len(possible_strs)):
            if str_ele != possible_strs[m]:
                new_str_l[l] = possible_strs[m]
                new_replace_str.append("".join(new_str_l))
            
    return new_replace_str

def str_deletion(new_str):
    x_l = list(new_str)
    for i in range(0,len(x_l)):
        x_l = list(new_str)
        del x_l[i]
        new_delete_str.append("".join(x_l))
    return new_delete_str


def str_insertion(new_str, possible_strs):
    global new_add_str
    poss_str_l = list(possible_strs)
    new_str_l = list(new_str)
    for m in poss_str_l:
        for n in range(len(new_str_l)+1):
            new_add_str.append(new_str_l[:n] + [m] + new_str_l[n:])

    new_add_str = [''.join(sublist) for sublist in new_add_str]
    
    return new_add_str



new_str = sample_str
str_replacement(new_str, possible_strs)
print("Strings with replacement are: ")
print(new_replace_str)
print(len(new_replace_str))
print("----------------------------------")

str_deletion(new_str)
print("The strings after deletion are: ")
print(new_delete_str)
print(len(new_delete_str))
print("----------------------------------")

str_insertion(new_str, possible_strs)
print("The strings after insertion are: ")
print(new_add_str)
print(len(new_add_str))
print("----------------------------------")

new_input_str = new_replace_str + new_delete_str + new_add_str

print("The complete string is ")
print(new_input_str)
print(len(new_input_str))
print("----------------------------------")


new_output = []
indices_mat_i = []
indices_mat_j = []

for i in range(0,len(new_input_str)):
    str_check = new_input_str[i]
#    print(str_check)
    count = 0
    for j in range(0,len(geno)):
        if str_check == geno[j]:
            count = count + 1
            indices_mat_i.append(i)
            indices_mat_j.append(j)
#            print(i, str_check, j, geno[j])
            new_output.append(phenoBinary[j])
    if count == 0:
        new_output.append("0")


arrlist = np.array(new_output) 
unique, counts = np.unique(arrlist, return_counts=True)
w = dict(zip(unique, counts))
#print(w)
#print("-----------------------------")

total_outputs = 0
for kv in w.items():
    total_outputs = total_outputs+kv[1]
#print("Total sum is ",total_outputs) 

for kv in w.items():
    file1.write(geno[ind])
    file1.write('\t')
    file1.write(str(phenoBinary[ind]))
    file1.write('\t')
    file1.write(str(kv[0]))
    file2.write(str(kv[0]))
    file1.write('\t')
    file1.write(str(kv[1]))
    file1.write('\t')
    file1.write(str(kv[1]/total_outputs))
    file3.write(str(kv[1]/total_outputs))
    file1.write("\n")
    file2.write("\n")
    file3.write("\n")
print("Bingo.")

file1.close()
file2.close()
file3.close()
file4.close()
#print(indices_mat_j )
#print(len(indices_mat_j ))
```
