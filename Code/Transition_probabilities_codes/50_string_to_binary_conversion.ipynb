```python
import numpy as np
from collections import OrderedDict
import random as rm
#from ttictoc import tic,toc
#t1 = tic()

file = open("HP_output_data_files/HP_25_173_possible_new_unique_str.txt","r")
file1 = open("HP_output_data_files/HP_25_173_possible_new_unique_str_binary.txt","w")


arr = file.read()
arr_t = arr.split("\n")
arr_t.pop()

print(arr_t)


mappings = {"U": "11", "D": "00", "R": "01","L": "10","A": "0"}

def strs_conversion(strings):
    converted_strs = []
    for string in strings:
        binary_string = ""
        for char in string:
            binary_string += mappings[char]
        converted_strs.append(binary_string)
    return converted_strs

converted_strs = strs_conversion(arr_t)

print(converted_strs)

for string in converted_strs:
        file1.write(string + "\n")

file.close()
file1.close()
```
