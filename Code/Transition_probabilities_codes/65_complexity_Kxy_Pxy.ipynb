```python
# For Umar
import numpy as np
#import LZ78Complexity
import KC # used for Lempel-Ziv complexity estimation
import matplotlib.pyplot as plt
import math as mt

file1 = open("data/unique_binary_str_n.txt","r")
file2 = open("data/prob_reverse_output_1.txt","r")

str_arr = file1.read()
str_arrlist = str_arr.split("\n")
str_arrlist.pop()

prob_arr = file2.read()
prob_arrlist = prob_arr.split("\n")
prob_arrlist.pop()

#print(str_arrlist[1])

#print(prob_arrlist)

# random SS
#BinaryString = ["01010","0101000","001011","11010"]
#print(BinaryString)
Z = []
for s in str_arrlist:
    k = KC.calc_KC(s)
    Z.append(np.round(k,1))

print(Z)

#upper_bound = []

#for i in range(0,len(Z)):
#    temp = 1/pow(2,float(Z[i]))
#    upper_bound.append(temp)

# sum of probabilities    
nplist = np.array(prob_arrlist)
print(nplist)


nplist1 = np.array(nplist)
#maxxx = max(nplist)
#minnn = min(nplist)

#print(maxxx)
#print(minnn)

final_prob = []
for i in range(0, len(nplist1)):
    q = nplist1[i]
    if float(q) > 0:
#        print("The value for ",i, "is",q)
        qlog = mt.log10(float(q))
#        print("The log value for ",i, "is",qlog)
        final_prob.append(qlog)
    else:
        final_prob.append(0)

#nplist1 = np.array(final_prob)   
    
#maxxx1 = np.max(nplist1)
#minnn1 = np.min(nplist1)

#print(maxxx1)
#print(minnn1)
    
print(final_prob)

plt.figure()
#plt.plot(Z1, upper_bound)
plt.scatter(Z,final_prob)
plt.title('conditional complexity vs Probability \n for x = 1')
plt.ylabel(r'$\log_{10} P(x \rightarrow y)$',fontsize=18)
plt.xlabel(r'$\tilde{K}(y|x)$',fontsize=18)
plt.rc('xtick',labelsize=15)
plt.rc('ytick',labelsize=15)
plt.tight_layout()
plt.savefig('s28_Pxy_Kxy.jpg',dpi=2500)
plt.show()



file1.close()
file2.close()
```
